## 1. Primitivas / Primitives

**ES** Calculemos la primitiva de la función  
**EN** Let us compute the primitive of the function

$$
f(x) = a x^2 + b x + c
$$

```python
from sympy import symbols, integrate
from sympy.abc import a, b, c, x

f = a*x**2 + b*x + c
primitive = integrate(f, x)
primitive
```

**ES** La primitiva general es  
**EN** The general primitive is

$$
\frac{a x^3}{3} + \frac{b x^2}{2} + c x + C
$$

## 2. Ecuaciones diferenciales / Differential Equations

### 2.1 EDO básica / Basic ODE

**ES** Resolvamos la ecuación diferencial  
**EN** Let us solve the differential equation

$$
y' = 6x^2 - 3x^2 y
$$

```python
import sympy as sp

x = sp.Symbol('x')
y = sp.Function('y')

eq = sp.Eq(y(x).diff(x), 6*x**2 - 3*x**2*y(x))
solution = sp.dsolve(eq, y(x))
solution
```

**ES** La solución general incluye una constante C1.  
**EN** The general solution includes a constant C1.

## 3. Integrales definidas y área / Definite integrals and area

### 3.1 Integral definida sencilla / Simple definite integral

**ES** Calcular  
**EN** Compute

$$
\int_0^{\pi/2} \cos(x)\,dx
$$

```python
from sympy import integrate, cos, pi

x = symbols('x')
result_def = integrate(cos(x), (x, 0, pi/2))
result_def
```

### 3.2 Área de un círculo / Area of a circle

**ES** Área de un círculo de radio 2 como integral de una función.  
**EN** Area of a circle of radius 2 as an integral of a function.

$$
A = \int_{-2}^{2} 2\sqrt{4-x^2}\,dx
$$

```python
from sympy import sqrt

x = symbols('x')
area_circle = integrate(2*sqrt(4 - x**2), (x, -2, 2))
area_circle, float(area_circle)
```

**ES** Resultado esperado: A = 4π.  
**EN** Expected result: A = 4π.

## 4. Integrales dobles / Double integrals

### 4.1 Área de un rectángulo / Rectangle area

**ES** El área de un rectángulo [a,b]×[c,d] es  
**EN** The area of a rectangle [a,b]×[c,d] is

$$
\int_a^b\int_c^d 1\,dy\,dx = (b-a)(d-c)
$$

```python
from sympy import simplify
from sympy.abc import a, b, c, d, x, y

I1 = integrate(1, (y, c, d))
rect_area = simplify(integrate(I1, (x, a, b)))
rect_area
```

### 4.2 Círculo como integral doble / Circle as double integral

**ES** Área del círculo de radio 2.  
**EN** Area of a circle of radius 2.

Forma cartesiana / Cartesian form:

$$
A = \int_{-2}^{2}\int_{-\sqrt{4-x^2}}^{\sqrt{4-x^2}} 1\,dy\,dx
$$

Forma polar / Polar form:

$$
A = \int_0^{2\pi}\int_0^2 r\,dr\,d\theta
$$

```python
from sympy import symbols, sqrt, pi

x, y = symbols('x y')

# Cartesian double integral
A_cart = integrate(
    integrate(1, (y, -sqrt(4-x**2), sqrt(4-x**2))),
    (x, -2, 2)
)
A_cart
```

```python
from sympy.abc import r, theta

A_polar = integrate(
    integrate(r, (r, 0, 2)),
    (theta, 0, 2*pi)
)
A_polar
```

## 5. Integral impropia / Improper integral

**ES** Ejemplo típico:  
**EN** Typical example:
