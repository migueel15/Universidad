```python
from sympy import Sum, symbols, oo

n = symbols("n", integer=True, nonnegative=True)

s = Sum((n + 3) / 2**n, (n, 0, oo))
print(s.doit())
print(s)
```

```python
from sympy import symbols, integrate, sin, cos

x = symbols("x")

result = integrate((sin(x) * cos(x)) / (2 + cos(x) ** 2), x)
print(result)
```

```python
from sympy import symbols, integrate, exp, oo

x = symbols("x")

result = integrate(exp(-(x**2)), (x, 0, oo))
print(result)
```

```python
from sympy import symbols, Limit, oo, sqrt

n = symbols("n", integer=True, positive=True)

a_n = sqrt(n**2 + n) - n

limit = Limit(a_n, n, oo)
result = limit.doit()

print(result)
```

```python
from sympy import symbols, sin, series

x = symbols("x")

result = series(sin(x), x, 0, 11).removeO()
taylor = result.coeff(x, 5)
print(result)
print(taylor)
```

```python
from sympy import integrate, pi, symbols

r, theta = symbols("r theta")

result = integrate(r**3, (r, 1, 2), (theta, 0, 2 * pi))
print(result)
```

```python
from sympy import Sum, symbols, oo

n = symbols("n", integer=True, nonnegative=True)

s = Sum((n + 3) / 2**n, (n, 0, oo))
print(s.doit())
```

```python
from sympy import symbols, sqrt, integrate

x = symbols("x")

f = sqrt(x)
g = x**2

area = integrate(f - g, (x, 0, 1))
print(area)
```

```python
import sympy as sp

x = sp.Symbol("x")
y = sp.Function("y")

ode = sp.Eq(x - y(x) - x * y(x).diff(x), 0)
sol = sp.dsolve(ode, ics={y(1): 1})

print(sol)
```

```python
import sympy as sp

x = sp.Symbol("x")
y = sp.Function("y")

ode = sp.Eq(-2 * x * y(x).diff(x) + y(x), x**2 + 1)
sol = sp.dsolve(ode, ics={y(1): 0})

print(sol)
```
