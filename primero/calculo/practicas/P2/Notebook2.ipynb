## 1. Sequences and Limits / Sucesiones y Límites

A sequence $\{a_n\}$ converges to $L$ if $\lim_{n\to\infty} a_n = L$.  
SymPy: `Limit(expr, n, oo).doit()`

### 💻 Code from slides / Código de las diapositivas

```python
# Converging sequence / Sucesión convergente: a_n = 1/n
a_n = 1 / n
lim1 = Limit(a_n, n, oo)
print('Limit of 1/n =', lim1.doit())   # 0

# Oscillating sequence / Sucesión oscilante: sin(n)
a_f = sin(n)
lim2 = Limit(a_f, n, oo)
print('Limit of sin(n) =', lim2.doit())  # AccumBounds(-1,1)
```

```python
# lim (n^2+1)/(2n^2-3)  — slide example
a_n = (n**2 + 1) / (2*n**2 - 3)
lim = Limit(a_n, n, oo)
print('Result:', lim.doit())   # 1/2
display(lim)
```

```python
# lim (sqrt(n^2+n) - n)  — rationalization trick / truco de racionalización
n_pos = symbols('n', positive=True)
a_n2  = sqrt(n_pos**2 + n_pos) - n_pos
print('Simplified:', simplify(a_n2))
print('Limit:', Limit(a_n2, n_pos, oo).doit())  # 1/2
```

6. $\lim_{n\to\infty}\sin\dfrac{\pi n-1}{2-3n}$
7. $\lim_{n\to\infty}\left(\sqrt[3]{n+1}-\sqrt[3]{n}\right)$
8. $\lim_{n\to\infty}\dfrac{n^n e^{-n}\sqrt{2\pi n}}{n!}$ *(Stirling)*

```python
n_p = symbols('n', positive=True)

exprs = [
    ('1. (n^2+1)/(2n^2-3)',      (n**2+1)/(2*n**2-3),                   n   ),
    ('2. n/e^n',                  n/exp(n),                               n   ),
    ('3. sqrt(n^2+n)-n',          sqrt(n_p**2+n_p)-n_p,                  n_p ),
    ('4. (-1)^n/n',               (-1)**n/n,                              n   ),
]

for label, expr, var in exprs:
    print(label, '=', Limit(expr, var, oo).doit())
```

## 2. Numerical Series / Series Numéricas

SymPy: `Sum(expr, (n, 0, oo)).doit()`

| Test / Criterio | Rule / Regla |
|---|---|
| Geometric / Geométrica | $\sum r^n$, $|r|<1$ ⟹ converges |
| D'Alembert | $\ell=\lim|a_{n+1}/a_n|$: $<1$ conv., $>1$ div. |
| Leibniz | $a_n\searrow 0$ ⟹ $\sum(-1)^n a_n$ converges |
| Comparison / Comparación | $0\le a_n\le b_n$, $\sum b_n$ conv. ⟹ $\sum a_n$ conv. |

### 💻 Code from slides / Código de las diapositivas

```python
# Σ 1/n! = e  (converges / converge)
sumn = Sum(1/factorial(n_int), (n_int, 0, oo))
print('Sum of 1/n! =', sumn.doit())   # E

# Harmonic / Armónica: Σ 1/n  (diverges / diverge)
summ = Sum(1/n_int, (n_int, 1, oo))
print('Harmonic series:', summ.doit())  # oo

display(sumn)
```

```python
# Geometric / Geométrica: Σ (2/3)^n
s = Sum(Rational(2,3)**n_int, (n_int, 0, oo))
print('Σ (2/3)^n =', s.doit())   # 3
print('Formula: 1/(1-2/3) =', 1/(1-Rational(2,3)))
display(s)
```

```python
# Exponential / Exponencial: Σ 2^n/n! = e^2
s2 = Sum(2**n_int/factorial(n_int), (n_int, 0, oo))
r2 = s2.doit()
print('Σ 2^n/n! =', r2)                    # exp(2)
print('Numerical:', float(r2))              # 7.389056
display(s2)
```

```python
# D'Alembert helper / Función auxiliar para D'Alembert
def dalembert(a_n_expr, var=n_int):
    ratio = simplify(a_n_expr.subs(var, var+1) / a_n_expr)
    L = Limit(Abs(ratio), var, oo).doit()
    if L < 1:   verdict = 'CONVERGES'
    elif L > 1: verdict = 'DIVERGES'
    else:       verdict = 'INCONCLUSIVE (L=1)'
    return L, verdict

# Test: Σ 2^n/n! — converges
L1, v1 = dalembert(2**n_int/factorial(n_int))
print(f'Σ 2^n/n!  →  L={L1}  →  {v1}')

# Test: Σ n^n/n! — diverges
L2, v2 = dalembert(n_int**n_int/factorial(n_int))
print(f'Σ n^n/n!  →  L={L2}  →  {v2}')
```

## 3. Taylor Series / Series de Taylor

$$f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(x_0)}{n!}(x-x_0)^n$$

SymPy: `series(f, x, x0, order)` — order is exclusive (e.g. `order=6` gives up to $x^5$).

| Function | Series | Radius |
|---|---|---|
| $e^x$ | $\sum x^n/n!$ | $\infty$ |
| $\sin x$ | $\sum(-1)^n x^{2n+1}/(2n+1)!$ | $\infty$ |
| $\cos x$ | $\sum(-1)^n x^{2n}/(2n)!$ | $\infty$ |
| $\ln(1+x)$ | $\sum(-1)^{n+1}x^n/n$ | $(-1,1]$ |
| $(1+x)^\alpha$ | binomial | $(-1,1)$ |

### 💻 Code from slides / Código de las diapositivas

```python
# e^x around x=0 (Maclaurin)
t_exp = series(exp(x), x, 0, 6)
print(t_exp)
# Remove Big-O / Quitar resto
print('Polynomial:', t_exp.removeO())
display(t_exp)
```

```python
# sin(x) up to degree 10
t_sin = series(sin(x), x, 0, 11)
print('Only odd powers, alternating signs!')
display(t_sin)
```

```python
# log(1+x)
t_log = series(log(1+x), x, 0, 8)
print('Convergence: |x| <= 1, x != -1')
display(t_log)
```

```python
# (1+x)^alpha — binomial series
t_binom = series((1+x)**alpha, x, 0, 5)
display(t_binom)
```

```python
# arcsinh(x) = sinh^(-1)(x)
t_arcsinh = series(asinh(x), x, 0, 8)
display(t_arcsinh)
```

### 📊 Taylor approximations of sin(x) / Aproximaciones de Taylor de sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_vals, np.sin(x_vals), 'k-', lw=2.5, label=r'$\sin(x)$ (exact)')

colors  = ['#e74c3c', '#e67e22', '#2ecc71', '#3498db']
orders  = [1, 3, 5, 9]
for order, color in zip(orders, colors):
    poly = series(sin(x), x, 0, order+1).removeO()
    f    = sp.lambdify(x, poly, 'numpy')
    y    = np.clip(f(x_vals), -3, 3)
    ax.plot(x_vals, y, '--', color=color, lw=1.5, label=f'$T_{{{order}}}(x)$')

ax.set(xlim=(-2*np.pi, 2*np.pi), ylim=(-2, 2),
       xlabel='x', ylabel='y',
       title='Taylor approximations of sin(x) / Aproximaciones de Taylor de sin(x)')
ax.legend(); ax.grid(alpha=0.3)
ax.axhline(0, color='gray', lw=0.5); ax.axvline(0, color='gray', lw=0.5)
plt.tight_layout(); plt.show()
```

### ✏️ Exercises / Ejercicios

**A) Compute Taylor series at $x_0=0$:**  
$\log(1+x)$, $(1+x)^\alpha$, $\sin x$, $\operatorname{arcsinh} x$

**B) Limits via Taylor / Límites mediante Taylor:**  
1. $\lim_{x\to0}\frac{e^x-1-x}{x^2}$ &nbsp;&nbsp;
2. $\lim_{x\to0}\frac{\cos x-1+x^2/2}{x^4}$ &nbsp;&nbsp;
3. $\lim_{x\to0}\frac{\sin x-x}{x^3}$ &nbsp;&nbsp;
4. $\lim_{x\to0}\frac{\log(1+x)-x+x^2/2}{x^3}$

**C) Taylor polynomial + error / Polinomio de Taylor + error:**  
Compute $T_4(x)$ of $\cos x$ at $x_0=0$. Approximate $\cos(0.1)$ and bound $|E_4(0.1)|$.

```python
print('=== A: Taylor series ===')
for name, f in [('log(1+x)', log(1+x)), ('(1+x)^alpha', (1+x)**alpha),
                ('sin(x)',   sin(x)),    ('arcsinh(x)',   asinh(x))]:
    print(f'\n{name}:'); display(series(f, x, 0, 7))
```

```python
print('=== B: Limits via Taylor ===')
lims = [
    ('(e^x-1-x)/x^2',            (exp(x)-1-x)/x**2),
    ('(cosx-1+x^2/2)/x^4',       (cos(x)-1+x**2/2)/x**4),
    ('(sinx-x)/x^3',              (sin(x)-x)/x**3),
    ('(log(1+x)-x+x^2/2)/x^3',   (log(1+x)-x+x**2/2)/x**3),
]
for name, expr in lims:
    print(f'  lim {name} = {Limit(expr, x, 0).doit()}')
```

```python
print('=== C: cos(x) — T4 and error bound ===')

T4 = series(cos(x), x, 0, 5)
T4_poly = T4.removeO()
print('T_4(x) =', T4_poly)
display(T4)

approx = float(T4_poly.subs(x, 0.1))
exact  = float(sp.cos(sp.Rational(1,10)).evalf())
bound  = (0.1)**5 / float(factorial(5))

print(f'\nT_4(0.1)  = {approx:.10f}')
print(f'cos(0.1)  = {exact:.10f}')
print(f'Error bound |E_4| <= {bound:.2e}')
print(f'Actual error        = {abs(exact-approx):.2e}')
print(f'Bound holds: {abs(exact-approx) <= bound}')
```

## 4. Field of Convergence / Campo de Convergencia

For $\sum a_n x^n$: radius $R = \lim_{n\to\infty}|a_n/a_{n+1}|$. Check endpoints separately.

```python
def radius_of_convergence(a_n_expr, var=n_int):
    """R = lim |a_n / a_{n+1}|  (ratio test for power series)"""
    ratio = Abs(a_n_expr / a_n_expr.subs(var, var+1))
    return Limit(ratio, var, oo).doit()

power_series = [
    ('Σ (-1)^n/n · x^n',         (-1)**n_int / n_int),
    ('Σ (x+2)^n/(3^n√(n+1))',    1 / (3**n_int * sqrt(n_int+1))),
    ('Σ n! · x^n',               factorial(n_int)),
    ('Σ x^n/(n·2^n)',            1 / (n_int * 2**n_int)),
]

print('=== Radius of convergence / Radio de convergencia ===')
for name, a in power_series:
    R = radius_of_convergence(a)
    print(f'  {name:35s}  R = {R}')
```

## 5. Series Derived from the Geometric / Series Derivadas de la Geométrica

```python
# Golden ratio via Fibonacci / Número áureo de Fibonacci
fibs = [1, 1]
for _ in range(18): fibs.append(fibs[-1]+fibs[-2])
ratios = [fibs[i+1]/fibs[i] for i in range(19)]
phi = (1+5**0.5)/2
print('Last 5 ratios:')
for i, r in enumerate(ratios[-5:], 15):
    print(f'  n={i}: {r:.10f}  error={abs(r-phi):.2e}')
print(f'φ = {phi:.10f}')
```

```python
# Leibniz formula for π / Fórmula de Leibniz para π
# π/4 = 1 - 1/3 + 1/5 - 1/7 + ...
def leibniz_pi(k): return 4*sum((-1)**i/(2*i+1) for i in range(k))

pi_exact = float(sp.pi)
print('Estimating π with Leibniz:')
for k in [10, 100, 1000, 10000]:
    est = leibniz_pi(k)
    print(f'  {k:6d} terms: π ≈ {est:.8f}  error={abs(est-pi_exact):.2e}')
```

```python
# Complex series / Series complejas
# Σ i^n/3^n  and  Σ e^(iπn/2)/2^n

s1 = Sum(I**n_int/3**n_int, (n_int, 0, oo))
r1 = sp.simplify(s1.doit())
print(f'Σ i^n/3^n = {r1}')

s2 = Sum(exp(I*pi*n_int/2)/2**n_int, (n_int, 0, oo))
r2 = sp.simplify(s2.doit())
print(f'Σ e^(iπn/2)/2^n = {r2} ≈ {r2.evalf()}')
```

## 📝 Quick Reference / Referencia Rápida

| Task / Tarea | SymPy |
|---|---|
| Sequence limit / Límite de sucesión | `Limit(expr, n, oo).doit()` |
| Infinite series / Serie infinita | `Sum(expr, (n, 0, oo)).doit()` |
| D'Alembert ratio | `Limit(abs(a_{n+1}/a_n), n, oo)` |
| Taylor series | `series(f, x, x0, order)` |
| Remove O() | `.removeO()` |
| Evaluate at point | `.subs(x, value)` |
| Simplify | `simplify(expr)` |
| Pretty print | `display(expr)` |

**Remember / Recuerda:**
1. `series(f, x, 0, 6)` gives terms up to $x^5$ (order 6 = exclusive upper bound).
2. `.removeO()` strips the $O(x^n)$ remainder term.
3. SymPy's `Limit` handles $\pm\infty$, one-sided limits, and complex expressions.
4. For power series, always check the **endpoints** of the convergence interval separately.
